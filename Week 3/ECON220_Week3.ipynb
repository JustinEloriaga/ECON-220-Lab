```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Next, I specify again the path and load the csv file

```python
# Writing the path
path = "/Users/justineloriaga/Documents/GitHub/ECON-220-Lab/Week 3/random_dataset.csv"

# Loading the data
data = pd.read_csv(path)
data
```

Suppose we want to know the proportion of Blue M&M's per pack. We can get that using the form $$blue\_prop = \frac{\text{\# of Blue M\&M Per Pack}}{\text{Total M\&M Per Pack}} $$

We can generate a variable using the simple variable creation method below

```python
data['Blue_prop'] = data['Blue'] / data['Total']
data
```

From here, we can generate a for loop to create variables for the other colors. In essence, instead of repeating the command 6 more times, we can generate a loop that will do it in one go.

```python
colors = ['Orange', 'Green', 'Yellow', 'Red', 'Brown', 'Other']
for color in colors:
    data[color + '_prop'] = data[color] / data['Total']
```

Next, let us plot a histogram. A histogram is a useful representation particularly for understanding distributions. Here, let us understand the distribution of the proportion of blue M&M's

```python
plt.figure()
data['Blue_prop'].hist(bins=20, color='blue')
plt.title('Histogram of Blue Proportion')
plt.xlabel('Proportion')
plt.ylabel('Frequency')
plt.show()
```

```python
mean_proportions = data[['Blue_prop', 'Orange_prop', 'Green_prop', 'Yellow_prop', 'Red_prop', 'Brown_prop', 'Other_prop']].mean()
mean_proportions
```

```python
variance_proportions = data[['Blue_prop', 'Orange_prop', 'Green_prop', 'Yellow_prop', 'Red_prop', 'Brown_prop', 'Other_prop']].var()
variance_proportions
```

Next, let us discuss some simple data management logistics, like sorting a dataset

```python
# Sort the data by the Total column
data.sort_values(by='Total', ascending=False)
```
